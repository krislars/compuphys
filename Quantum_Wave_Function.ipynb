```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
dx=0.04
dx2=dx*dx
k0=5.5*np.pi
dt=dx2/20.
xmax=6.
xs=np.arange(-xmax,2.*xmax+dx/2,dx)
```

```python
psr=np.exp(-0.5*(xs/0.5)**2)*np.cos(k0*xs)
psi=np.exp(-0.5*(xs/0.5)**2)*np.sin(k0*xs)
V=np.zeros(xs.size)
V[xs>3]=200.0
plt.plot(xs,psr,label='real')
plt.plot(xs,psi,label='imag')
plt.plot(xs,np.sqrt(psr**2+psi**2),label='prob')
plt.plot(xs,V/np.max(V),'k')
plt.title('t = 0')
plt.legend()
```

```python
N=3000
for t in range(N):
    psr[1:-1]=psr[1:-1] - (dt/dx2)*(psi[2:] + psi[:-2] - 2*psi[1:-1]) + dt*V[1:-1]*psi[1:-1]
    psi[1:-1]=psi[1:-1] + (dt/dx2)*(psr[2:] + psr[:-2] - 2*psr[1:-1]) - dt*V[1:-1]*psr[1:-1]
plt.plot(xs,psr,label='real')
plt.plot(xs,psi,label='imag')
plt.plot(xs,np.sqrt(psr**2+psi**2),label='prob')
plt.plot(xs,V/np.max(V),'k')
plt.title('t = '+str(round(N*dt,3))+' s')
plt.legend()
```
