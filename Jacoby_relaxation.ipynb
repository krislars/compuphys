```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Set boundaries: 1V at edges
a=np.ones([5,5])
# Initialize with guess:
a[1:4,1:4]=0
a
```

```python
# Set boundaries: 1V at edges
anew=np.ones([5,5])
# Initialize with guess:
anew[1:4,1:4]=0
anew
```

```python
# Rerun this cell to watch the relaxation work
for i in np.arange(1,4):
    for j in np.arange(1,4):
        anew[i,j]=(a[i,j-1]+a[i,j+1]+a[i-1,j]+a[i+1,j])/4.
anew
```

```python
for i in np.arange(1,4):
    for j in np.arange(1,4):
        a[i,j]=(a[i,j-1]+a[i,j+1]+a[i-1,j]+a[i+1,j])/4.
a
```

```python
fig, ax = plt.subplots()
ax.contour(a)
fig.show()
```

```python
x = np.arange(5)
y = np.arange(5)
x, y = np.meshgrid(x, y)
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_wireframe(x,y,a)
fig.show()
```
